```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

mu, sigma = 2, 0.5
v = np.random.normal(mu,sigma,10000)

fig = plt.figure(figsize = (7,5))
ax = fig.gca()

count, bins, ignored = ax.hist(v, bins=50, normed=1)


ax.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
               linewidth=2, color='r')
ax.add_patch(
    patches.Arrow(
        1.2,            # x
        0.8,            # y
        0.5,            # dx
        0.0,            # dy
        width=0.1,      # optional - defaults to 1.0
        facecolor="red"
    )
)

ax.text(2.5, 0.7, r'$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{- \frac{(x-\mu)^2}{2\sigma^2}}$', fontsize=12)

plt.savefig('normal.pdf')
plt.savefig('normal.eps')
plt.savefig('normal.png', dpi=150)
plt.show()
```
